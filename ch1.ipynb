````python
import pandas as pd
from google.colab import drive

# Google Drive 마운트
drive.mount('/content/drive')

# data.csv 파일 경로 (실제 파일 경로에 맞게 수정해주세요)
# 예: /content/drive/MyDrive/data.csv
# 또는 특정 폴더 안에 있다면: /content/drive/MyDrive/MyFolder/data.csv
file_path = '/content/drive/MyDrive/military_meal_2026_H1.csv'

try:
    # CSV 파일 불러오기
    df = pd.read_csv(file_path)

    # 데이터 확인
    print("데이터 상위 5행:")
    print(df.head())

    print("\n데이터 정보:")
    df.info()

    print("\n결측치 확인:")
    print(df.isnull().sum())

except FileNotFoundError:
    print(f"오류: 파일이 존재하지 않습니다. 경로를 확인해주세요: {file_path}")
except Exception as e:
    print(f"오류 발생: {e}")
````
